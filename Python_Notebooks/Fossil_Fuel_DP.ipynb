```python
import pandas as pd
import numpy as np
```

```python
# importing the Fossil_Fuel_Data.csv file

df = pd.read_csv('Fossil_Fuel_Data.csv',skiprows=4)
```

```python
#Setting the index to Country Name column

df.set_index('Country Name')
```

```python
df.columns
```

```python
# Dropping the columns
df.drop(columns=['Country Code','Indicator Name','Indicator Code'],inplace=True)
```

```python
year_list = []

for n in range(1960,1981):
    year_list.append(str(n))
```

```python
df.drop(columns=year_list,axis=1,inplace=True)
```

```python
df.drop(columns=['2015','2016','2017','2018','2019','2020','2021'],inplace=True)
```
