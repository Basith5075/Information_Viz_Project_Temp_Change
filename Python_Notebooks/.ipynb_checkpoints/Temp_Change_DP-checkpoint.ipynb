```python
import pandas as pd
import numpy as np
```

```python
# Loading the file into the python notebook
df = pd.read_csv('Temperature_Change_Data.csv',encoding='latin-1')
```

```python
display(df)
display(df.columns)
```

```python
# Renaming the Area Column to 'Country_Name' 
df.rename(columns={'Area':'Country_Name'},inplace=True)

# Setting Country_Name as default index

df.set_index('Country_Name')

# Filtering the data

df = df[df ['Element']=='Temperature change']
#df = df[df['Months']=='Meteorological year']


display(df.head(50))
```

```python
# drop unwanted columns from df
columns_deleted = ['Area Code','Area Code (M49)','Months Code','Unit','Element Code','Element']

df.drop(columns= columns_deleted,inplace=True)

#display(df.head(50))

year_list = []

year_list = ['Y1961F','Y1962F','Y1963F','Y1964F']

for n in range(1961,2022):
    year = 'Y'+str(n)+'F'
    year_list.append(year)

df.drop(columns=year_list,inplace=True)

year_list_1 = []

for n in range(1961,1981):
    year = 'Y'+str(n)
    year_list_1.append(year)

for n in range(2015,2022):
    year='Y'+str(n)
    year_list_1.append(year)
#display(year_list_1)

df.drop(columns=year_list_1,inplace=True)

display(df.head(50))
```

```python
df_task2 = df[df['Months']=='Meteorological year']
```

```python
df_task2.drop(columns='Months',inplace=True)
df_task2
df_task2.to_csv('Task2_cleaned.csv')
```
